## 크롤링

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# requsts.get()으로 url정보 요청하기 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
(KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}

url = 'https://www.koreabaseball.com/Record/Team/Pitcher/Basic1.aspx'

res = requests.get(url, headers=headers)
```

```python
res
```

```python
# BeautifulSoup()으로 웹페이지 분석 준비하기
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# 기록 정보가 담긴 태그와 클래스 찾기
len(soup.select("table.tData"))
# len(soup.find('div', {'class':'record_result'}))
```

```python
tr_elements = soup.select("table.tData > tbody > tr")
```

```python
for e in tr_elements:
  print(e.text)
```

```python
# 데이터 긁어오기
total_data_list = []
for e in tr_elements:
  td_elements = e.select("td")
  data_list = []
  for td_e in td_elements:
    data_list.append(td_e.text.strip())
  total_data_list.append(data_list)
```

```python
total_data_list
```

```python
df = pd.DataFrame(total_data_list,
                  columns = ['No', 'TEAM', 'ERA', 'G', 'W', 'L', 'SV', 'HLD', 'WPCT', 'IP', 'H', 'HR', 'BB', 'HBP', 'SO', 'R', 'ER', 'WHIP']
                )
df
```
