## ë°ì´í„° ì „ì²˜ë¦¬ ê³¼ì • í•¨ìˆ˜ë¡œ ë§Œë“¤ê¸°

```python
# í•¨ìˆ˜ ìƒì„±
def create_kto_data(yy, mm):
    # 1. ë¶ˆëŸ¬ì˜¬ Excel íŒŒì¼ ê²½ë¡œë¥¼ ì§€ì •
    file_path = './files/kto_{}{}.xlsx'.format(yy, mm)
    
    # 2. Excel íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
    df = pd.read_excel(file_path, header=1, usecols='A:G', skipfooter=4)
    
    # 3. "ê¸°ì¤€ì—°ì›”" ì»¬ëŸ¼ ì¶”ê°€
    df['ê¸°ì¤€ì—°ì›”'] = '{}-{}'.format(yy, mm)
    
    # 4. "êµ­ì " ì»¬ëŸ¼ì—ì„œ ëŒ€ë¥™ ì œê±°í•˜ê³  êµ­ê°€ë§Œ ë‚¨ê¸°ê¸°
    continents_list = ['ì•„ì‹œì•„ì£¼', 'ë¯¸ì£¼', 'êµ¬ì£¼', 'ëŒ€ì–‘ì£¼', 'ì•„í”„ë¦¬ì¹´ì£¼', 'ê¸°íƒ€ëŒ€ë¥™', 'êµí¬ì†Œê³„'] # ì œê±°í•  ëŒ€ë¥™ëª…
    df_country = df[df['êµ­ì '].isin(continents_list) == False]
    df_country = df_country.reset_index(drop=True)

    # 5. "ëŒ€ë¥™" ì»¬ëŸ¼ ì¶”ê°€
    continents = ['ì•„ì‹œì•„']*25 + ['ì•„ë©”ë¦¬ì¹´']*5 + ['ìœ ëŸ½']*23 + ['ëŒ€ì–‘ì£¼']*3 + ['ì•„í”„ë¦¬ì¹´']*2 + ['ê¸°íƒ€ëŒ€ë¥™'] + ['êµí¬']    # ëŒ€ë¥™ ì»¬ëŸ¼ ìƒì„±ì„ ìœ„í•œ ëª©ë¡ ë§Œë“¤ì–´ ë†“ê¸°
    df_country['ëŒ€ë¥™'] = continents
    
    # 6. êµ­ê°€ë³„ "ê´€ê´‘ê°ë¹„ìœ¨(%)" ì»¬ëŸ¼ ì¶”ê°€
    df_country['ê´€ê´‘ê°ë¹„ìœ¨(%)'] = round(df_country['ê´€ê´‘'] / df_country['ê³„'] * 100, 1)
    df_country['ìœ í•™ìƒë¹„ìœ¨(%)'] = round(df_country['ìœ í•™/ì—°ìˆ˜'] / df_country['ê³„'] * 100, 1)
    
    
    # 7. "ì „ì²´ë¹„ìœ¨(%)" ì»¬ëŸ¼ ì¶”ê°€(ì „ì²´ ê´€ê´€ê° ìˆ˜ ì¤‘ êµ­ê°€ë³„ ë¹„ìœ¨)
    tourist = df_country['ê´€ê´‘'].sum()
    df_country['ê´€ê´‘ê°_ì „ì²´ë¹„ìœ¨(%)'] = round(df_country['ê´€ê´‘'] / tourist * 100, 1)
    
    foreign_student = df_country['ìœ í•™/ì—°ìˆ˜'].sum()
    df_country['ìœ í•™ìƒ_ì „ì²´ë¹„ìœ¨(%)'] = round(df_country['ìœ í•™/ì—°ìˆ˜'] / foreign_student * 100, 1)
    
    
    # 8. ê²°ê³¼ ì¶œë ¥
    return(df_country)
```

```python
# create_kto_data() í•¨ìˆ˜ë¥¼ í™œìš©í•´ì„œ 2018ë…„ 12ì›” ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df_test = create_kto_data(2018, 12)
df_test.head()
```

```python
# ìœ í•™ìƒ ë¹„ìœ¨ì´ ê°€ì¥ ë§ì€ ë‚˜ë¼ëŠ”?
df_test.sort_values(by='ìœ í•™ìƒ_ì „ì²´ë¹„ìœ¨(%)', ascending=False).head()
```
